```python
#importing the libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#taking the dataset
source = "http://bit.ly/w-data"
data = pd.read_csv(source)
print("data imported")
data.head(13)
```

```python
#plotting the dataset
data.plot(x = 'Hours',y = 'Scores',style ="o")
plt.title('Hours v/s Scores')
plt.xlabel("hours studied")
plt.ylabel("scores in percentage")
plt.show()
```

```python
#dividing the data into inputs and outputs
X = data.iloc[:,:-1].values
y = data.iloc[:,1].values
```

```python
#splitting the traning and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state =0)
```

```python
#training the dataset with linear regression model
from sklearn.linear_model import LinearRegression
Regressor = LinearRegression()
Regressor.fit(X_train,y_train)
print("training successful")

```

```python
#plotting the regressor line and the testing data
line = Regressor.coef_*X+Regressor.intercept_
plt.scatter(X,y)
plt.plot(X,line)
plt.show()

```

```python
#printing the testing dataset
print(X_test)
```

```python
#making the predicting variable
y_pred = Regressor.predict(X_test)
```

```python
#comparing the testing dataset with the predicting dataset
df =pd.DataFrame({'y_test':y_test,"y_pred":y_pred})
df
```

```python
#testing the given data of hours = 9.25
hours =9.25
hour  =np.array(hours).reshape(-1,1)
ans_pred = Regressor.predict(hour)
print("no. of hours ={}".format(hours))
print("predicted answer ={}".format(ans_pred[0]))
```

```python
#calculating the mean absolute error
from sklearn.metrics import mean_absolute_error
print("mean absolute error ={}".format(mean_absolute_error(y_test,y_pred)))
```
